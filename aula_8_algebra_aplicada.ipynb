```python
from sklearn.datasets import load_iris
import pandas as pd

dataset = load_iris()

df = pd.DataFrame(
    data=dataset.data,
    columns=dataset.feature_names
)
df['class'] = dataset.target
df['class'] = df['class'].map(
    lambda x: dataset.target_names[x])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='class')
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components = 2, 
            perplexity   = 40, 
            n_iter       = 5000)

new_data = tsne.fit_transform(dataset.data)
```

```python
new_df = df = pd.DataFrame(
    data=new_data,
    columns=['dimension 1', 'dimension 2']
)
new_df['class'] = dataset.target
new_df['class'] = df['class'].map(
    lambda x: dataset.target_names[x])
```

```python
sns.scatterplot(data=new_df, x="dimension 1", y="dimension 2", hue="class")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

new_data = pca.fit_transform(dataset.data)
```

```python
new_df = df = pd.DataFrame(
    data=new_data,
    columns=['component 1', 'component 2']
)
new_df['class'] = dataset.target
new_df['class'] = df['class'].map(
    lambda x: dataset.target_names[x])

sns.scatterplot(data=new_df, x="component 1", y="component 2", hue="class")
plt.show()
```
