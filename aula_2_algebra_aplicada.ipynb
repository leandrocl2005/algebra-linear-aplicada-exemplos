```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# lendo o arquivo
M = Image.open('family.jpg')
# convertendo em matriz
M = np.asarray(M, dtype=np.float32)/255
# visualizando
plt.imshow(M, aspect='auto')
plt.axis('off')
plt.show()
```

```python
M = Image.open('pine.jpg')
N = Image.open('sunset.png')
M = np.asarray(M, dtype=np.float32)/255
N = np.asarray(N, dtype=np.float32)/255
P = 0.5 * M + 0.5 * N
plt.imshow(P, aspect='auto')
plt.axis('off')
plt.show()
```

```python
#  https://stackoverflow.com/questions/65344928/sepia-filter-inverting

# lendo o arquivo
M = Image.open('pine.jpg')
# convertendo em matriz
M = np.asarray(M, dtype=np.float32)/255
R = 0.393 * M[:,:,0] + 0.769 * M[:,:,1] + 0.189 * M[:,:,2]
G = 0.89 * R
B = 0.69 * R
R = R.clip(0,1); G = G.clip(0,1); B = B.clip(0,1)
M[:,:,0] = R; M[:,:,1] = G; M[:,:,2] = B
plt.imshow(M, aspect='auto')
plt.axis('off')
plt.show()
```

```python
#  https://www.codementor.io/@isaib.cicourel/intermediate-image-filters-mj6y7abx4

# lendo o arquivo
M = Image.open('pine.jpg')
# convertendo em matriz
M = np.asarray(M, dtype=np.float32)/255
R = 0.759 * M[:,:,0] + 0.398 * M[:,:,1] + 0.194 * M[:,:,2]
G = 0.676 * M[:,:,0] + 0.354 * M[:,:,1] + 0.173 * M[:,:,2]
B = 0.524 * M[:,:,0] + 0.277 * M[:,:,1] + 0.136 * M[:,:,2]
R = R.clip(0,1); G = G.clip(0,1); B = B.clip(0,1)
M[:,:,0] = R; M[:,:,1] = G; M[:,:,2] = B
plt.imshow(M, aspect='auto')
plt.axis('off')
plt.show()
```
