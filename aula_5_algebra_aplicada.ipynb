```python
set_largura = [1.4 , 1.3 , 1.5 , 1.4 , 1.7 , 1.6]
set_comprim = [0.2 , 0.2 , 0.2 , 0.3 , 0.4 , 0.2]
ver_largura = [4.7 , 4.5 , 4.9 , 4.0 , 4.6 , 4.5]
ver_comprim = [1.4 , 1.5 , 1.5 , 1.3 , 1.5 , 1.3]
vir_largura = [6.0 , 5.1 , 5.9 , 5.6 , 5.8 , 6.6]
vir_comprim = [2.5 , 1.9 , 2.1 , 1.8 , 2.2 , 2.1]
new_largura = [4.9]
new_comprim = [1.7]
```

```python
plt.scatter(set_largura , set_comprim , c='r')
plt.scatter(ver_largura , ver_comprim , c='g')
plt.scatter(vir_largura , vir_comprim , c='b')
plt.scatter(new_largura , new_comprim , c='m')
plt.show()
```

```python
plt.scatter(set_largura , set_comprim , c='r',
           label="setosa")
plt.scatter(ver_largura , ver_comprim , c='g',
           label="versicolor")
plt.scatter(vir_largura , vir_comprim , c='b',
           label="virginica")
plt.scatter(new_largura , new_comprim , c='m',
            marker='*',
            s=120,
            label="nova")
plt.legend()
plt.title("Espcies de Iris",
         fontweight='bold',
         fontsize=14)
plt.xlabel("comprimento da ptala",
          fontsize=12)
plt.ylabel("comprimento da spala",
          fontsize=12)
plt.show()
```
