### Transformação linear via expressão

```python
def T(nupla):
  x1 = nupla[0] # primeira coordenada
  x2 = nupla[1] # segunda coordenada
  return [2 * x1 - x2, - x1 + 3 * x2]
```

```python
import numpy as np

v = np.array([2,3])
T(v)
```

### Transformação Linear via matrizes

```python
def T(v):
  A = np.array([[2, -1],[-1, 3]])
  return np.matmul(A, v)
```

```python
v = np.array([2,3])
T(v)
```

### Transformando conjunto de dados

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
dataset = load_iris()
x = dataset.data[:50,0]
y = dataset.data[:50,1]
tx = T(dataset.data[:50,:2].T)[0,:50]
ty = T(dataset.data[:50,:2].T)[1,:50]
plt.scatter(tx, ty, label="transformados")
plt.scatter(x, y, label="originais")
plt.title("Amostras de Iris Setosa")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.legend()
plt.show()
```

### Exemplo de rotação

```python
theta = np.deg2rad(45)

A = np.array([
    [np.cos ( theta ), np.sin ( theta )],
    [-np.sin ( theta ), np.cos ( theta )]
])

intervalo = np.arange(0,1,0.05) 

x, y = np.meshgrid(intervalo, intervalo)
tx = A[0][0] * x + A[0][1] * y
ty = A[1][0] * x + A[1][1] * y
plt.scatter(x, y, alpha=0.3)
plt.scatter(tx, ty, alpha=0.3)
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.show()
```

### Exemplo de cisalhamento

```python
a = 1.1

A = np.array([[1, a], [0, 1]])

intervalo = np.arange(0,1,0.05) 

x, y = np.meshgrid(intervalo, intervalo)
tx = A[0][0] * x + A[0][1] * y
ty = A[1][0] * x + A[1][1] * y
plt.scatter(x, y, alpha=0.3)
plt.scatter(tx, ty, alpha=0.3)
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.show()
```
