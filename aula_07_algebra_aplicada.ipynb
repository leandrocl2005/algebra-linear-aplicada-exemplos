# Gráficos tridimensionais

```python
from sklearn.datasets import load_iris
dataset = load_iris()
```

```python
print("Nº de linhas:", dataset.data.shape[0])
print("Nº de colunas:", dataset.data.shape[1])
print("Variáveis", dataset.feature_names)
```

```python
import pandas as pd
df = pd.DataFrame(
    data=dataset.data,
    columns=dataset.feature_names
)
df['class'] = dataset.target
df['class'] = df['class'].map(
    lambda x: dataset.target_names[x])
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (5,5))
ax = Axes3D(fig)

x = df['sepal length (cm)']
y = df['sepal width (cm)']
z = df['petal width (cm)']
colors = df['class'].map({
    'virginica': 'blue',
    'setosa': 'green',
    'versicolor': 'red'
})

ax.scatter(x, y, z, c=colors, alpha = 0.7)
ax.set_xlabel('sepal length (cm)')
ax.set_ylabel('sepal width (cm)')
ax.set_zlabel('petal width (cm)')
plt.show()
```

```python
import numpy as np
def f(x, y):
    try:
        return (x ** 2 - y ** 2 ) / (x ** 2 + y ** 2)
    except:
        return np.nan
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure(figsize = (5,5))
ax = Axes3D(fig)

x = np.arange(-2, 2, 0.01)
y = np.arange(-2, 2, 0.01)
x, y = np.meshgrid(x, y)
z = f(x,y)

surf = ax.plot_surface(x, y, z, linewidth=0, antialiased=False, alpha=0.3)

# limites para o eixo z
ax.set_zlim(-2, 2)
# número de marcações no eixo z
ax.zaxis.set_major_locator(LinearLocator(5))
# arredondamento de z em 2 casas decimais
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
# ângulo de visualização
ax.view_init(30,20) 
plt.show()
```
