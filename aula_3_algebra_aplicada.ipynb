```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# lendo o arquivo
M = Image.open('monk.jpg')
# convertendo em matriz
M = np.asarray(M, dtype=np.float32)/255
# verificando dimensões
M.shape
```

```python
# criando uma identidade de ordem 426
P = np.eye(426)
# zerando as 142 primeiras linhas de I
P[:142,:] = 0
# zerando as últimas 142 linhas de I
P[426-142:,:] = 0
```

```python
# Multiplicando as camadas de M por R
M[:,:,0] = np.matmul(P, M[:,:,0])
M[:,:,1] = np.matmul(P, M[:,:,1])
M[:,:,2] = np.matmul(P, M[:,:,2])
```

```python
# visualizando
plt.imshow(M, aspect='auto')
plt.axis('off')
plt.show()
```

```python
B = Image.open('boy.jpg')
G = Image.open('girl.jpg')
B = np.asarray(B, dtype=np.float32)/255
G = np.asarray(G, dtype=np.float32)/255
print(B.shape, G.shape)
```

```python
P = np.eye(854) 
P[122:244,:] = 0 
P[366:488,:] = 0
P[610:732,:] = 0
I = np.eye(854)
```

```python
for i in [0, 1, 2]:
  B[:,:,i] = np.matmul(P, B[:,:,i])
  G[:,:,i] = np.matmul(I - P, G[:,:,i])
```

```python
plt.imshow(B + G, aspect='auto')
plt.axis('off')
plt.show()
```
