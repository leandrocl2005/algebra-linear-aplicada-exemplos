```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

M = Image.open('boy.png').convert('RGB')
M = np.asarray(M, dtype=np.float32)/255
plt.imshow(M, aspect='auto')
plt.show()
```

```python
print(M.shape)
```

```python
M[200,300,:]
```

```python
for i in range(0,620):
  for j in range(0,600):
    if M[i,j,0] == 0.49411765 \
      and M[i,j,1] == 0.12156863 \
      and M[i,j,2] == 0.12156863:
      M[i,j,0] = 0
      M[i,j,1] = 0
      M[i,j,2] = 1
```

```python
plt.imshow(M, aspect='auto')
plt.show()
```

```python
def d(x,y):
  return np.linalg.norm(x - y)
```

```python
y = np.array([
  0.49411765,
  0.12156863,
  0.12156863
])
for i in range(0,620):
  for j in range(0,600):
    x = np.array([
        M[i,j,0],
        M[i,j,1],
        M[i,j,2]
    ])
    if d(x,y) < 0.01:
      M[i,j,0] = 0
      M[i,j,1] = 0
      M[i,j,2] = 1
```

```python
plt.imshow(M, aspect='auto')
plt.show()
```

```python
import imageio
M = 255 * M
M = M.astype(np.uint8)
imageio.imwrite('boy_blue.jpg',M)
```
