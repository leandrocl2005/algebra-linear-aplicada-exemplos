```python
t = np.linspace(0, 2 * np.pi, 100)
x = 3 * np.cos(t)
y = 2 * np.sin(t)
plt.plot(x, y)
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.show()
```

```python
theta = np.linspace(0, 2 * np.pi, 1000)
r = np.cos(2 * theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.plot(x, y)
plt.xlim([-1.1, 1.1])
plt.ylim([-1.1, 1.1])
plt.show()
```
