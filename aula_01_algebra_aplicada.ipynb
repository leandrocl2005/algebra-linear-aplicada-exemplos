## N-uplas reais

O conjunto ${\mathbb{R}}^n$ de <b>$n$-uplas reais</b> é definido por:
$${\mathbb{R}}^n :=\left\{(u_1,u_2,\ldots,u_n); u_i \in {\mathbb{R}}\ \forall i=1,2,\ldots,n \right\}.$$

Sejam $u = (u_1,\ldots,u_n)$ e $v=(v_1,\ldots,v_n)$ $n$-uplas em ${\mathbb{R}}^n$. A <b>soma usual</b> de $u$ com $v$ é definida por:
$$u+v := (u_1+v_1,\ldots,u_n+v_n).$$

Sejam $\alpha \in {\mathbb{R}}$ e  $u = (u_1,\ldots,u_n) \in {\mathbb{R}}^n$. A <b>multiplicação por escalar usual</b> de $\alpha$ por $u$ é definida por:
$$\alpha . u = (\alpha . u_1, \ldots, \alpha . u_n).$$

## Exemplo teórico

Sejam $u = (-1,0,2,5)$, $v = (2,0,0,9)$ 4-uplas e $\alpha = 2$ escalar real. Calcule $\alpha . u + v$.

Resposta: $(0,0,4,19)$

## N-uplas com Python

Existem várias formas de representar uma n-upla com Python. Utilizaremos <b>numpy arrays</b>. Para isso lembre-se de importar, antes de tudo, a biblioteca numpy:

```python
import numpy as np
```

Para representar a $n$-upla $u = (-1,0,2,5)$ fazemos:

```python
u = np.array([-1,0,2,5])
```

Para representar a $n$-upla $v = (2,0,0,9)$ fazemos:

```python
v = np.array([2,0,0,9])
```

Escalares reais não inteiros podem ser aproximados por números flutuantes. Por exemplo, para representar o escalar $\alpha = \sqrt{2}$ fazemos:

```python
alpha = 1.41421356
```

Escalares inteiros podem ser representados por objetos do tipo inteiro. Por exemplo, para $\alpha = 2$ fazemos:

```python
alpha = 2
```

<b>Observação:</b> Definimos a variável `alpha` duas vezes acima. Quando isso ocorre, apenas a última definição permanece. Logo, `alpha = 2`

Uma das vantagens de representar $n$-uplas utilizando numpy arrays é que as operações de soma e multiplicação por escalar já vêm prontas. Caso queremos calcular $\alpha . u + v$ fazemos:


```python
alpha * u + v
```

## Exemplo prático

Uma empresa deseja lançar a folha de pagamento bimestral de seus funcionários. São ao todo 4 funcionários que receberam em janeiro R\\$ 2560, R\\$ 2460, R\\$ 3000, R\\$ 3100 e R\\$1000. Já em fevereiro receberam R\\$ 2500, R\\$ 2400, R\\$ 3000, R\\$ 2900 e R\\$ 900 respectivamente. Sabendo que foi dado um aumento de 10% em janeiro, quanto seriam os lançamentos bimestrais (relativos aos meses de janeiro e fevereiro) respectivos para cada funcionário?

```python
# 5-upla representando pagamentos em janeiro
janeiro_pgts = np.array([2560,2460,3100,3000,1000])
# 5-upla representando pagamentos em fevereiro
fevereiro_pgts = np.array([2500,2400,3000,2900,900])
# escalar representando aumento de janeiro
aumento = 1.1
# Calculo da 5-upla representando pagamentos bimentrais relativos a janeiro e feveiro
bimestral_pgts = aumento * janeiro_pgts + fevereiro_pgts
# Mostrando pagamentos bimestrais
print(bimestral_pgts)
```

## Matrizes reais

O <b>conjunto de matrizes com $m$ linhas e $n$ colunas</b> sobre ${\mathbb{R}}$ é definido por:
$${\cal{M}}_{m\times n} := \left\{\left(a_{ij}\right); a_{ij}\in {\mathbb{R}}\ \forall i, j =1,\ldots,n\right\}.$$

Sejam $A = \left(a_{ij}\right),\ B = \left(b_{ij}\right) \in {\cal{M}}_{m\times n}$. Definimos a <b>soma usual</b> de $A$ com $B$ por:
$$A+B := \left(a_{ij} + b_{ij}\right).$$

Sejam $A = \left(a_{ij}\right) \in {\cal{M}}_{m\times n}$ e $\alpha \in {\mathbb{R}}$. Definimos a <b>multiplicação por escalar usual</b> de $\alpha$ por $A$ da seguinte forma:
$$\alpha . A := \left(\alpha.a_{ij}\right).$$

Seja $A = (a_{ij}) \in {\cal{M}}_{m\times n}$. A <b>transposta</b> da matriz $A$, denotada por $A^{T}$, é definida por $$A^{T} := \left(a_{ji}\right),$$ com $i=1,\ldots,m$ e $j=1,\ldots,n$. Note que $A^{T} \in {\cal{M}}_{n\times m}$.

## Exemplo teórico:

Sejam $\alpha = 2$ e $A, B \in {\cal{M}}_{3\times 2}$ com 
$$A=\left(\begin{array}{cc}
1 & 1 \\ 2 & 0 \\ 0 & 1
\end{array}\right)\ \text{e}\ 
B=\left(\begin{array}{ccc}
7 & 1 & 1 \\ 0 & -3 & 1
\end{array}\right).$$
Calcule $\alpha . A + B^{T}$. 

<b>Resposta:</b> $$\left(\begin{array}{cc}
9 & 2 \\ 5 & -3 \\ 1 & 3
\end{array}\right)$$

## Matrizes com Python

Para representar matrizes em Python utilizaremos novamente numpy arrays. Como já importamos antes a biblioteca numpy, não precisamos importá-la novamente.

Para representar a matriz $A$ do exemplo anterior, fazemos:

```python
A = np.array([ [1,1],[2,0],[0,1]])
```

Para representar a matriz B do exemplo anterior, fazemos:

```python
B = np.array([[7,1,1], [0,-3,1]])
```

A vantagem de representar matrizes como numpy arrays se dá pelo fato de haver várias operações já pré-definidas. Por exemplo, para calcular a transposta da matriz B fazemos:

```python
B.T
```

Vamos agora resolver o exemplo teórico anterior usando Python:

```python
alpha = 2
alpha * A + B.T
```
